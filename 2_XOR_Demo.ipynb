An MLP is especially effective for problems that are not linearly separable. A classic example is the XOR function.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# True data
x = 2
y_true = 6
```

```python
# Initial weight
w = 1.0
eta = 0.05
epochs = 10
history = []
```

```python
for epoch in range(epochs):
    y_pred = w * x
    loss = (y_pred - y_true) ** 2
    grad = 2 * (w * x - y_true) * x
    w = w - eta * grad
    history.append((w, loss))

print(f"Final weight: {w:.3f}")
print(f"Final loss: {loss:.3f}")
```

```python
# Plot loss over time
weights, losses = zip(*history)
plt.figure(figsize=(6,4))
plt.plot(range(epochs), losses, marker='o')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss over Epochs (One Weight Model)")
plt.grid(True)
plt.show()
```
